```python
import pandas as pd
import matplotlib as plt
import numpy as np
import bls
```

```python
#get wage data (year/quarter and percentage)
wages = bls.get_series('CIU2020000000000A')
wages
```

```python
#get CPI data
cpi = bls.get_series('CUUR0000SA0')

#convert to csv format
cpi.to_csv('cpi_data.csv')

#read in CSV
cpi_data = pd.read_csv('cpi_data.csv')

#add DATE and CPI percentage column names to the CSV
cpi_data.columns = 'Date','CPI'
cpi_data
```

```python
# CPI data is monthly, Wages data is quarterly
# replace every 3 months with the corresponding quarter
cpi_data['Date'] = cpi_data['Date'].str.replace('-03', 'Q1')
cpi_data['Date'] = cpi_data['Date'].str.replace('-06', 'Q2')
cpi_data['Date'] = cpi_data['Date'].str.replace('-09', 'Q3')
cpi_data['Date'] = cpi_data['Date'].str.replace('-12', 'Q4')

# drop remaining rows that still have a '-' in the date column
newcpi_data = cpi_data[(cpi_data.Date.str.contains('-') == False)]

#save to a new CSV
newcpi_data.to_csv('cpi-clean.csv')
```

```python
# convert wages to CSV and read in a data frame
wages.to_csv('bls_wages_data_csv')
df = pd.read_csv('bls_wages_data_csv')

# rename columns
df.columns = 'Date','Wages'
df.head()
newdf = df
```

```python
newnum = 230.28
def process_wages(x):
    global newnum
    if type(x) is str:
        return x
    elif x:
        #quarterly asjustment and convert to percentage
        # use first wage value as baseline
        # CPI value adjusted by the wage-growth percentage
        newnum = (x/400) * newnum + newnum
        return newnum
    else:
        return
newwages_data = newdf.applymap(process_wages)
newwages_data
```

```python
# merge the two frames (cpi and wages)
merged_data = pd.merge(newcpi_data, newwages_data, on='Date')
merged_data
```

```python
# plot the frame
ax = merged_data.plot(kind='line', figsize=(20, 7))
ax.set_xticklabels(merged_data.Date, rotation=45)
ax.set_ylabel('CPI vs. Wages and salaries - 12-month change')
ax.set_xlabel('Dates')
ax.grid()
```
